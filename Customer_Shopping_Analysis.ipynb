```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping = {
    'Daily': 1,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df = df.where(pd.notnull(df), None)

for i, row in df.iterrows():
    sql = """
    INSERT INTO customer_shopping_behavior (
        customer_id, age, gender, item_purchased, category, purchase_amount,
        location, size, color, season, review_rating, subscription_status,
        shipping_type, discount_applied, previous_purchases,
        payment_method, frequency_of_purchases
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    values = tuple(None if pd.isna(x) else x for x in row[
        ['customer_id', 'age', 'gender', 'item_purchased', 'category', 'purchase_amount',
         'location', 'size', 'color', 'season', 'review_rating', 'subscription_status',
         'shipping_type', 'discount_applied', 'previous_purchases',
         'payment_method', 'frequency_of_purchases']
    ])
    cursor.execute(sql, values)

conn.commit()
print("✅ All CSV data inserted successfully into MySQL!")
```
